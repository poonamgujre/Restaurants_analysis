```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV into pandas
df = pd.read_csv('Dataset .csv')
df
```

```python
# top five row 
df.head()
```

```python
# all infomation about the dataset
df.info()
```

```python
print(df['Cuisines'])
```

```python
# check null value
df.isnull().sum()
```

```python
#fill missing values
df['Cuisines'] = df['Cuisines'].fillna(df['Cuisines'].mode()[0])
df.isnull().sum()
```

```python
total_cuisines =df['Cuisines'].str.split(', ').explode('cuisines ').value_counts()
top_cuisines = total_cuisines.head(3)
print("Top three cuisines are:",top_cuisines)
```

```python
barwidth =0.25
fig = plt.figure(figsize = (10, 5))
sns.barplot(x= top_cuisines.index, y =top_cuisines.values,palette ="Paired",hue =top_cuisines).set_title('Top three most common cuisines in the dataset')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
total_restaurants = len(df)
percentage = round((top_cuisines/total_restaurants)*100,2)
print("percentage of restaurants that serve each of the top cuisines","\n",percentage)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.index, y= top_cuisines.values,palette ="Paired",hue = percentage).set_title('percentage of Top three  cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
city_count = df['City'].value_counts()
number_of_restro = city_count.head()
print('city with the highest number of restaurants:' , city_count.idxmax())
print(number_of_restro)
```

```python
rating_by_each_city = df.groupby('City')['Aggregate rating'].mean()
print('The average rating for restaurants in each city','\n',rating_by_each_city)
```

```python
average_rating_by_city = rating_by_each_city.idxmax()
rating = rating_by_each_city.max()
print('The city with the highest average rating:',average_rating_by_city, '\n','Rating:',rating)
```
