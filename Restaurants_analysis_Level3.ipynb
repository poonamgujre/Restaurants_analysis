```python
# import Natural language tool kit library
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 
from collections import Counter
from nltk.corpus import wordnet
import re
```

```python
 #download some NLTK datasets and models
import nltk
nltk.download('vader_lexicon')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
stopwords = nltk.corpus.stopwords.words('english') # download stop word from nltk 
print(stopwords)
```

```python
# make an instance of SentimentIntensityAnalyzer
sia=SentimentIntensityAnalyzer()
```

```python
df.columns
```

```python
text_reviews = df['Rating text'].unique() # display all unique value from the column
text_reviews
```

```python
# initialize counters for keywords and lists for reviews
positive_keywords = Counter()
negative_keywords = Counter()
positive_reviews = []
negative_reviews = []
neutral_reviews = []
```

```python
#Process reviews
for review in text_reviews:
    sentiment_scores = sia.polarity_scores(review)
    compound_score = sentiment_scores['compound']
    tokens = re.findall(r'\b\w+\b', review.lower())
    filtered_tokens = [token for token in tokens if token not in stopwords]
    if compound_score > 0.05: 
        positive_reviews.append(review)
        positive_keywords.update(filtered_tokens)
    elif compound_score < -0.05: 
         negative_reviews.append(review)
         negative_keywords.update(filtered_tokens)
    else:  
         neutral_reviews.append(review)
```

```python
print("Positive Reviews:")
for review in positive_reviews:
    print(f"- {review}")

print("\nNegative Reviews:")
for review in negative_reviews:
    print(f"- {review}")

print("\nNeutral Reviews:")
for review in neutral_reviews:
    print(f"- {review}")

# Display the most common keywords
print("\nMost Common Positive Keywords:")
for keyword, count in positive_keywords.most_common(10):
    print(f"{keyword}: {count}")

print("\nMost Common Negative Keywords:")
for keyword, count in negative_keywords.most_common(10):
    print(f"{keyword}: {count}")
```

```python
df_analyze=df[['Rating text','Aggregate rating']].copy()
df_analyze['Review Length']=df_analyze['Rating text'].apply(lambda x: len(str(x)))
average_length_of_reviews=df_analyze.groupby('Aggregate rating')['Review Length'].mean()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=average_length_of_reviews.index, y=average_length_of_reviews.values)
plt.title('Average Review Length For Each Rating Category')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average review length character')
plt.xticks(rotation=45)
plt.show()
```
